```python
import sweetviz as sv
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
diabetes_report = sv.analyze(df)
diabetes_report.show_html('Diabetes_report.html')
```

```python
df1 = sv.compare(df[102:], df[:102])
df1.show_html('Compare.html')
```
